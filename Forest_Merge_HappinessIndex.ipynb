```python
#code to loop through each catagory and calculate distance from US average
#code to assign a score based on the above
#code to add all new rows for each county for index score


# TA: Making a copy of the dataframe to perserve the original plus deleting extra/irrelevant columns
happinessindex_df = health_forest_fips.copy()


del happinessindex_df['Mapping Zones']
del happinessindex_df['State_y'] # Redundant columns
del happinessindex_df['County_y']
del happinessindex_df['% < 18']
del happinessindex_df['% 65 and over']
del happinessindex_df['Unnamed: 0'] # This is just the index number
del happinessindex_df['# Households with Severe Cost Burden'] 
del happinessindex_df['% Severe Housing Cost Burden']

happinessindex_df = happinessindex_df.rename(columns = {"State_x" : "State", "County_x" : "County"})

# TA: List of columns

health_forest_fips.columns

# Data to be calculated:
# ~~~~~~~~~~~~~~~~~~~~~~~~
# % Unemployed
# Association Rate
# Life Expectancy
# % Frequent Mental Distress
# % Frequent Physical Distress
# % Homeowners
```

```python
happinessindex_df["% Unemployment Score"] = 0
# Column index 28
happinessindex_df["Association Rate Score"] = 0
# Column index 29
happinessindex_df["% Mentally Calm Score"] = 0
# Column index 30
happinessindex_df["% Physically Comfortable Score"] = 0
# Column index 31
happinessindex_df["% Homeowners Score"] = 0
# Column index 32
happinessindex_df["Life Expectancy Score"] = 0
# Column index 33

# del happinessindex_df["% Mentally Comfortable Score"]

for i in range(len(happinessindex_df)):
    
    employment_rate = 100 - happinessindex_df.iloc[i, 5]
    happinessindex_df.iloc[i, 28] = employment_rate/10
    
    happinessindex_df.iloc[i, 29] = happinessindex_df.iloc[i, 8]/10
    
    mentallycalm = 100 - happinessindex_df.iloc[i, 12]
    happinessindex_df.iloc[i, 30] = mentallycalm/10
    
    physicallycomfortable = 100 - happinessindex_df.iloc[i, 11]
    happinessindex_df.iloc[i, 31] = physicallycomfortable/10

    happinessindex_df.iloc[i, 32] = happinessindex_df.iloc[i, 19]/10
    

    happinessindex_df.iloc[i, 33] = happinessindex_df.iloc[i, 10]/10





    
happinessindex_df.head(len(happinessindex_df))
```

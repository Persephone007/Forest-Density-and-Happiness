```python
#  Dependencies
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
suicide_data = "2016_suicide.csv"
suicide_data_df = pd.read_csv(suicide_data)
suicide_data_df.head()
```

```python
forest_data = "forest_fips.csv"
forest_data_df = pd.read_csv(forest_data)
forest_data_df.head()
```

```python
forest_suicide_df = (pd.merge(forest_data_df,suicide_data_df, on=['FIPS']))
forest_suicide_df
```

```python
plt.scatter(forest_suicide_df["Tree Canopy Cover (m2/person)"],forest_suicide_df["Deaths"], 
           label = "Suicides vs Tree Canopy Cover (m2/person)")
plt.title("Suicides vs Tree Canopy Cover (m2/person)")
plt.xlabel("Canopy Cover (m2/person)")
plt.ylabel("# of Suicides")
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

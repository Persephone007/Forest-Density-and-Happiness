```python
#Dependencies 
%matplotlib inline 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
forest_file = "Resources/forest_fips.csv"
most_data_file = "Resources/full_data_csv.csv"
age_file = "Resources/Health_Data_CSV.csv"
```

```python
forest_file_df = pd.read_csv(forest_file)
most_data_file_df = pd.read_csv(most_data_file)
age_file_df = pd.read_csv(age_file)

```

```python
data_forest_college_df = pd.merge(forest_file_df, most_data_file_df, on="FIPS", how="outer")
first_merge_table_df = pd.merge(data_forest_college_df, age_file_df, on="FIPS", how="outer")
first_merge_table_df
```

```python
merge_table_df = first_merge_table_df[["FIPS","Tree Canopy Cover (m2/person)","% Some College","Income Ratio","Violent Crime Rate",
                                      "% < 18","% 65 and over"]]
merge_table_df.head()
```

```python
plt.scatter(merge_table_df["Tree Canopy Cover (m2/person)"],merge_table_df["Violent Crime Rate"], 
           label = "Crime Rate vs Tree Canopy Cover (m2/person)")
plt.title("Crime Rate vs Tree Canopy Cover (m2/person)")
plt.xlabel("Canopy Cover (m2/person)")
plt.ylabel("Violent Crime Rate")
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

```python
plt.scatter(merge_table_df["Tree Canopy Cover (m2/person)"],merge_table_df["Income Ratio"], 
           label = "Crime Rate vs Tree Canopy Cover (m2/person)")
plt.title("Income Ratio vs Tree Canopy Cover (m2/person)")
plt.xlabel("Canopy Cover (m2/person)")
plt.ylabel("Income Ratio")
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

```python
plt.scatter(merge_table_df["Tree Canopy Cover (m2/person)"],merge_table_df["% < 18"], 
           label = "Crime Rate vs Tree Canopy Cover (m2/person)")
plt.title("% < 18 vs Tree Canopy Cover (m2/person)")
plt.xlabel("Canopy Cover (m2/person)")
plt.ylabel("% < 18")
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

```python
plt.scatter(merge_table_df["Tree Canopy Cover (m2/person)"],merge_table_df["% 65 and over"], 
           label = "Crime Rate vs Tree Canopy Cover (m2/person)")
plt.title("% 65 and over vs Tree Canopy Cover (m2/person)")
plt.xlabel("Canopy Cover (m2/person)")
plt.ylabel("% 65 and over")
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

```python
plt.scatter(merge_table_df["Tree Canopy Cover (m2/person)"],merge_table_df["% Some College"], 
           label = "Crime Rate vs Tree Canopy Cover (m2/person)")
plt.title("% Some College and over vs Tree Canopy Cover (m2/person)")
plt.xlabel("Canopy Cover (m2/person)")
plt.ylabel("% Some College and over")
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

```python
NJ_file = "Resources/NJ_Specific_CSV.csv"
NJ_file_df = pd.read_csv(NJ_file)
NJ_file_df.head()
```

```python
NJ_data_df = NJ_file_df[["FIPS","State","County","% Fair/Poor","Mentally Unhealthy Days",
                                      "% Some College","% Unemployed"]]

# NJ_column_df = NJ_data_df.rename(columns = {"Unnamed: 0":"FIPS","Unnamed: 1":"State","Unnamed: 2":"County","Unnamed: 109":"% Unemployed"})

NJ_df = NJ_data_df.drop([0])
NJ_df
```

```python
NJ_final_df =NJ_df.dropna()
NJ_final_df
```

```python
NJ_forest_df = pd.merge(NJ_final_df,forest_file_df, on="FIPS", how="outer")
NJ_merge = NJ_forest_df.dropna()
NJ_merge
```

```python
NJ_merge_final = NJ_merge[["FIPS","Tree Canopy Cover (m2/person)","% Some College","% Fair/Poor","% Unemployed",
                                      "Mentally Unhealthy Days"]]
NJ_merge_final
```

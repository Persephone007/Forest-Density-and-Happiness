```python
import plotly.figure_factory as ff
import numpy as np
import pandas as pd
from matplotlib import cm
import matplotlib.pyplot as plt
import colorlover as cl
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
ranking_data_to_load = (r"C:\Users\lecha\Desktop\Forrest_Project\Charlene\NJ_Rankings_1.csv")
ranking_data = pd.read_csv(ranking_data_to_load)
```

```python

df_sample_r = ranking_data[ranking_data['State'] == 'New Jersey']

values = df_sample_r['County Rank'].tolist()
fips = df_sample_r['FIPS'].tolist()
endpts = list(np.mgrid[min(values):max(values)])

colorscale = ["#3B2853","#7550A6","#9E85C1","#C8B9DB","#E3DCED",
              "#963267","#E766B9","#FAAEF3","#F8BAE5","#FDEAF7",
              "#C7EEFF","#9CCAE9","#4CA0D7","#276894","#086AE9",
              "#020288","#E6E6E6","#B3B3B3","#808080","#5A5A5A",
              "#333333","#333333","#333333"]


fig = ff.create_choropleth(
    fips=fips, values=values, scope=['County Rank'], show_state_data=True,
    colorscale=colorscale, round_legend_values=True,
    plot_bgcolor='rgb(229,229,229)',
    paper_bgcolor='rgb(229,229,229)',
    legend_title='NJ Wellness Index Rankings - Worst to Best',
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.4},
    exponent_format=False,
)
fig.layout.template = None
fig.show()
```

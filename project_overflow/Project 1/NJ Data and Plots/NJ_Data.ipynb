```python
#Dependencies 
%matplotlib inline 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as n
forest_data = "forest_fips.csv"
full_data = "full_data_csv.csv"
age_data = "Health_Data_CSV.csv"
forest_data_df = pd.read_csv(forest_data)
full_data_df = pd.read_csv(full_data)
age_data_df = pd.read_csv(age_data)
```

```python
NJ_data = "NJ_Specific_CSV.csv"
NJ_data_df = pd.read_csv(NJ_data)
NJ_data_df.head()
```

```python
forest_college_df = pd.merge(forest_data_df,full_data_df, on="FIPS")
forest_college_df.head()
```

```python
NJ_forest_college_df = pd.merge(NJ_data_df,forest_college_df, on="FIPS")
NJ_forest_college_df
```

```python
plt.scatter(NJ_forest_college_df["Tree Canopy Cover (m2/person)"],NJ_forest_college_df["% Some College_y"], 
           label = "% Some College vs Tree Canopy Cover (m2/person) in NJ")
plt.title("% Some College vs Tree Canopy Cover (m2/person) in NJ")
plt.xlabel("Canopy Cover (m2/person)")
plt.ylabel("% Some College")
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

```python
fig.savefig('college.jpg')
```

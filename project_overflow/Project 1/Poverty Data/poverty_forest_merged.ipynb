```python
#  Dependencies
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
poverty_data = "census_poverty.csv"
poverty_data_df = pd.read_csv(poverty_data)
poverty_data_df.head()
```

```python
forest_data = "forest_fips.csv"
forest_data_df = pd.read_csv(forest_data)
forest_data_df.head()
```

```python
forest_poverty_df = (pd.merge(forest_data_df,poverty_data_df, on=['FIPS']))
forest_poverty_df
```

```python
plt.scatter(forest_poverty_df["Tree Canopy Cover (m2/person)"],forest_poverty_df["Poverty %"], 
           label = "Poverty vs Tree Canopy Cover (m2/person)")
plt.title("Poverty vs Tree Canopy Cover (m2/person)")
plt.xlabel("Canopy Cover (m2/person)")
plt.ylabel("Poverty %")
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

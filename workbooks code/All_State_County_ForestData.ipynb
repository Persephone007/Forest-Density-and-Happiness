```python
# This function is supposed to make the final dataset. It takes a list of statenames as the function
# arguement. The names correspond to the names of the files in the "State_Data" folder located in this repo.
# It drops the first three rows as those are just table headers.
# It also adds statenames in a seperate column in order know which counties belong to which states.
# More code will be added that renames the columns on the final dataframe but that will be implemented later.

def makeCountryTable(statenames):
       
    df_final = pd.DataFrame()
    
    for i in range(len(statenames)):
        
        try:

                
                # Making first dataframe
            name = statenames[i]
            filepath = "State_Data/" + name + "_Tables.xls"

            df1 = pd.ExcelFile(filepath)
            
            print(f"Iteration: {i}")
            print(f"Sheet Names for df1: {df1.sheet_names}")
        
                
            df1_data = df1.parse('7')
            
            df1_data = df1_data.drop(0)
            df1_data = df1_data.drop(1)
            df1_data = df1_data.drop(2)

            df1_data['State Name'] = statenames[i]
            
            df_final = df_final.append(df1_data)
                
                

#             # Making second dataframe
#             secondname = statenames[i+1]
#             secondfilepath = "State_Data/" + secondname + "_Tables.xls"

#             df2 = pd.ExcelFile(secondfilepath)
#             print(f"Sheet Names for df2: {df2.sheet_names}")

                
#             df2_data = df2.parse('7')
#             df1_data['State Name'] = statenames[i+1]


#             df_temp = df1_data.append(df2_data)

#             df_final = df_final.append(df_temp)

#             print(f'Making dataframe iteration: {i} statename: {statenames[i]}')
            
        except IndexError:
            
            break
            
    df_final = df_final.rename(columns = {"Unnamed: 1" : "Mapping Zones",
                                          "Unnamed: 2" : "Impervious Cover %",
                                          "Unnamed: 3" : "Impervious Cover (m2/person)",
                                          "Unnamed: 4" : "Tree Canopy Cover %",
                                          "Unnamed: 5" : "Tree Canopy Cover (m2/person)",
                                          "Unnamed: 6" : "Total Green Space (ha)",
                                          "Unnamed: 7" : "Canopy Green Space %",
                                          "Unnamed: 8" : "Available Green Space (ha)"})
    
    return df_final


answer = makeCountryTable(state_names)

answer.head(len(answer))
```

```python
# This is a function to count the total rows for each state dataset. The total number of rows will be less
# in the final dataframe as the code that makes the final dataframe drops 3 rows per state dataset.

def countingrows(states): # is the list object state_names
    
    statecount = 0
    count = 0
    
    for state in states:
        
        filepath = "State_Data/" + state + "_Tables.xls"
        
        temp = pd.ExcelFile(filepath)
        
        temp = temp.parse('7')
    
        statecount += len(temp)
        
        count += 1
        
        print(count)
        
    
    return statecount

print(f"Total rows are: {countingrows(state_names)}")
```
